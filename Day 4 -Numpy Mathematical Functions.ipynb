### Numpy Mathematical Functions
    In Numpy,  Mathematical FunctionsNumPy provides a wide range of mathematical functions such as arithmetic operations, trigonometric functions, exponential functions, logarithmic functions, and statistical calculations. These functions help perform complex calculations easily without writing long loops in Python.
    
    Using NumPy mathematical functions improves performance, readability, and efficiency when working with large datasets in data analysis, machine learning, and scientific computing.

```python
import numpy as np
```

```python
arr1 = np.arange(1,10).reshape(3,3)
arr2 = np.arange(1,10).reshape(3,3)
print(arr1)
print(arr2)
```


### Addition of Two Numpy Array 

```python
arr1 + arr2
```

```python
np.add(arr1, arr2)
```

### Subtraction of Two NumPy Array

```python
arr1 - arr2
```

```python
np.subtract(arr1, arr2)
```

### Division of Two NumPy Array

```python
arr1 / arr2
```

```python
np.divide(arr1, arr2)
```

### Multiplication of Two NumPy Array

```python
arr1 * arr2
```

```python
np.multiply(arr1 , arr2)
```

### Matrix Product of Two NumPy Array (matrix)


```python
arr1 @ arr2
```

```python
arr1.dot(arr2)
```
