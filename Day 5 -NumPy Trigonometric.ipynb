### NumPy Trigonometric Functions 
    NumPy Trigonometric Functions are mathematical functions in the NumPy library used to calculate trigonometric values such as sine, cosine, tangent, and their inverse functions. These functions operate efficiently on NumPy arrays, allowing element-wise calculations on large datasets.
    In NumPy, trigonometric functions take angles in radians as input and return the corresponding trigonometric values. They are widely used in scientific computing, engineering, signal processing, physics simulations, and data analysis.
    
    Common NumPy trigonometric functions include:

    - np.sin() – Calculates the sine of an angle
    - np.cos() – Calculates the cosine of an angle
    - np.tan() – Calculates the tangent of an angle


```python
import numpy as np
```

### np.sin()

#### 1. Sine value of the angle in degree

```python
sin_90 = np.sin(90)
print("sine value of angle 90 in degree = ", sin_90)
```

#### 2. Sine value of the angle in radians 

```python
np.pi = 3.14
sin_90 = np.sin(90 * np.pi/180 )
print("sine value of angle 90 in radians = ",sin_90)
```

### np.cos()

#### 1. Cosine value of angle in degrees 

```python
cos_180 = np.cos(180)
print("Cosine value of angle 180 in degree = ", cos_180)
```

#### 2. Cosine value of the angle in radians

```python
cos_180 = np.cos(180 * np.pi/180)
print ("cosine value of angle 180 in radians = ",cos_180)
```

### np.tan()

#### 1. Tangent value of angle in degrees

```python
tan_60 = np.tan(60)
print("Tangent value of angle 60 in degree = ",tan_60)
```

#### 2. Tangent value of angle in radians

```python
tan_60 = np.tan(60 * np.pi/180)
print("Tangent value of angle 60 in radians = ", tan_60)
```

### Graphical Representation of Trigonometric sine Function 

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,3 * np.pi, 0.2)
y = np.sin(x)
plt.plot(x,y)
plt.title ("Graphical Representation of sine function")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

### Graphical Representation of Trigonometric Cosine Function 

```python
x = np.arange(0,3 * np.pi, 0.2)
y = np.cos(x)
plt.plot(x,y)
plt.title ("Graphical Representation of cosine function")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

### Graphical Representation Of Trigonometric Tangent Function

```python
x = np.arange(0,3 * np.pi, 0.2)
y = np.tan(x)
plt.plot(x,y)
plt.title ("Graphical Representation of tangent function")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
