## Python NumPy String Operations Methods
    NumPy provides a special module called numpy.char that allows you to perform vectorized string operations on NumPy arrays. These methods work similarly to Python string functions but are optimized to operate on entire arrays of strings at once, making them faster and more efficient for data processing.

    -> Some commonly used NumPy string methods include:

    - np.char.add() – Concatenate two strings
    - np.char.multiply() – Repeat strings
    - np.char.upper() / np.char.lower() – Change letter case
    - np.char.split() – Split strings
    - np.char.join() – Join strings
    - np.char.replace() – Replace text
    - np.char.strip() – Remove spaces

```python
import numpy as np
ch_name = "Code Block Academy"
str1 = "Leaning Python Numpy"
```

### np.char.add()

```python
add_string = np.char.add(ch_name, str1)
print("Adding two string ch_name and str1 : ",add_string)
```

### np.char.multiply()

```python
duplicate_ch_name = np.char.multiply(ch_name,2)
print("Duplicated ch_name :",duplicate_ch_name)
```

### np.char.capitalize()

```python
cap_ch_name = np.char.capitalize(ch_name)
print("Capitalize ch_name :",cap_ch_name)
```

### np.char.lower()

```python
low_ch_name = np.char.lower(ch_name)
print("Lower ch_name :",low_ch_name)
```

### np.char.upper()

```python
upp_ch_name = np.char.upper(ch_name)
print("Upper ch_name : ",upp_ch_name)
```

### np.char.title()

```python
tit_upp_ch_name = np.char.title(upp_ch_name)
print("Title upp_ch_name :",tit_upp_ch_name)
```

### np.char.center()

```python
cen_ch_name = np.char.center(ch_name, 30, fillchar = "*")
print("Center ch_name :",cen_ch_name)
```

### np.char.split()

```python
spl_ch_name = np.char.split(ch_name)
print("Splite spl_ch_name :",spl_ch_name)
```

### np.char.splitlines()

```python
splitlines_str = np.char.splitlines("Indian\nAi\nProduction")
print("Splitlines : ",splitlines_str)
```

### np.char.join()

```python
join_dmy = np.char.join([":","/"],["dmy", "dmy"])
print("joint dmy by : and / :",join_dmy)
```

### np.char.replace

```python
rep_ch_name = np.char.replace(ch_name, "Block", " Binary Lock") 
print("Replace old string by new = ", rep_ch_name) 
```

### np.char.equal() 

```python
equ_str = np.char.equal(ch_name, str1) 
print("Is ch_name equal to str1 = ", equ_str )
```

### np.char.not_equal()

```python
not_equ_str = np.char.equal(ch_name, str1) 
print("Is ch_name not equal to str1 = ", equ_str ) 
```

### np.char.count() 


```python
count_A = np.char.count(ch_name, "A") 
print("Character 'A' present ", count_A , " times in a string - ", ch_name) 
```

### np.char.find()

```python
find_Block = np.char.count(ch_name, "Block") 
print("String  'Block' found at index ", find_Block , " from string - ", 
ch_name)  
```

### np.char.isalpha()

```python
str_isalpha = np.char.isalpha("Hello")  
print("Is String 'Hello' alphabet - ", str_isalpha )
```

### np.char.isdecimal()

```python
str_isdecimal = np.char.isdecimal("Hello")  
print("Is String 'Hello' decimal - ", str_isdecimal )
```

### np.char.isdigit()

```python
str_isdigit = np.char.isdigit("Hello")  
print("Is String 'Hello' digit - ", str_isdigit )
```

### np.char.islower()

```python
str_islower = np.char.islower("Hello")  
print("Is String 'Hello' lower - ", str_islower )
```

### np.char.isupper()

```python
str_isupper = np.char.isupper("Hello World")  
print("Is String 'Hello India' uppercase - ", str_isupper )
```

### np.char.isnumeric()

```python
str_isnumeric = np.char.isnumeric("Hello")  
print("Is String 'Hello' numeric - ", str_isnumeric )
```

### np.char.isspace()

```python
str_isspace = np.char.isspace("Hello World")
print ("Is Strint 'Hello India' space - ",str_isspace )
```
