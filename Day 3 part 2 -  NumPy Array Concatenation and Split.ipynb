## Python Numpy Array concatenation and split
    NumPy provides powerful functions to combine (concatenate) multiple arrays and divide (split) arrays into smaller parts. These operations are very useful in data processing, machine learning, and numerical analysis when handling large datasets.

```python
import numpy as np
```

```python
arr1 = np.arange(1,17).reshape(4,4)
print (arr1)
```

```python
arr2 = np.arange(17,33).reshape(4,4)
print (arr2)
```

### Concatenate list1 and list2

```python
list1 = [2,4,6,7]
list2 = [1,8,3,5]
list1 + list2
```

### Concatenate of arr1 and arr2

```python
np.concatenate((arr1,arr2))
```

```python
print(arr1)
print(arr2)
```

###  Concatenate array row wise using axis attribute 

```python
np.concatenate((arr1,arr2), axis = 1)
```

###  Vertical order concatenate in arrays

```python
np.vstack((arr1,arr2))
```

###  Horizontal order concatenate in arrays 

```python
np.hstack((arr1,arr2))
```

```python
np.hstack((arr1,arr2,arr1))
```

### Break the Array using Split() method

```python
np.split(arr1, 2)
```

```python
np.split(arr1, 2, axis=1)
```
