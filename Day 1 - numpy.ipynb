#  Numpy 
    NumPy (Numerical Python) is a powerful open-source Python library used for numerical computing and working with large, multi-dimensional arrays and matrices efficiently.

    It provides:
    
    - Fast mathematical operations
    - Multi-dimensional array object (ndarray)
    - Built-in mathematical and statistical functions
    - Broadcasting for efficient computations
    - Support for linear algebra operations

### Convert multidimensional array in one dimensional - flatten()

```python
import numpy as np 

arr_name = np.array([[1,2,3],[4,5,6,]])
array_name = arr_name.ravel()
print("Using ravel :",array_name)

```

```python
import numpy as np
arr_name = np.array([[1,2,3],[4,5,6,]])
array_name = arr_name.flatten()
print("Using flatten :",array_name)
```
