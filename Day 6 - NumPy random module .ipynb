## Python NumPy random module
    Using the NumPy random module, we can generate random integers, floating-point numbers, random samples from distributions, and shuffled arrays efficiently. It is faster and more powerful than Python’s built-in random module when working with large datasets and numerical computations.
    
    The random module is available through numpy.random, which includes many useful functions such as rand(), randint(), choice(), permutation(), and seed() to generate and control randomness.

```python
import numpy as np
import random
```

```python
rd_num = np.random.random(1)
rd_2d_array = np.random.random((3,3))
print (rd_num)
print (rd_2d_array)
```

### np.random.randint()

```python
rd_no = np.random.random(1)
rd_2D_arr = np.random.randint(1,4, (4,4)) 
rd_3D_arr = np.random.randint(1,4, (2,4,4)) 
print (rd_no)
print (rd_2D_arr)
print (rd_3D_arr)
```

### np.random.seed()

```python
np.random.seed(10) 
rd_3D_arr = np.random.randint(1,4, (2,4,4)) 
print(rd_3D_arr)
```

### np.random.rand()

```python
arr_2D = np.random.rand(3,3) 
print(arr_2D)
```

### np.random.randn()

```python
arr_2D = np.random.randn(3,3) 
print(arr_2D)
```

### np.random.choice() 

```python
x = [1,2,3,4] 
choice_from_x = np.random.choice(x)  
print(choice_from_x ) 
```

```python
for i in range(15): 
    print(np.random.choice(x)) 
```

### np.random.permutation() 

```python
x_permute = np.random.permutation(x) 
print(x_permute)
```
