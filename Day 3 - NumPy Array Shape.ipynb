## Python NumPy array shape Function

    In NumPy, provides the shape function to check the dimensions of an array. It tells us how many rows, columns, or axes an array has.
    Shape means the structure of the array.

```python
import numpy as np
```

#### The shape of a NumPy 1D array

```python
arr_1D = np.array([1, 2, 3])
print("One dimensional NumPy Array : \n",arr_1D)
shape_of_arr_1D = np.shape(arr_1D)
print("Shape of arr_1D = ",shape_of_arr_1D)
```

#### The shape of a Numpy 2D array 

```python
arr_2D = np.array([[1, 2, 3], [1, 2, 3], [1, 2, 3]])
print("One dimensional NumPy Array : \n",arr_2D)
shape_of_arr_2D = np.shape(arr_2D)
print("Shape of arr_2D = ",shape_of_arr_2D)
```

#### The shape of a Numpy 3D array

```python
arr_3D = np.array([[[1, 2, 3], [1, 2, 3], [1, 2, 3]], [[1, 2, 3], [1, 2, 3], 
[1, 2, 3]], [[1, 2, 3], [1, 2, 3], [1, 2, 3]]])
print ("Three dimensional NumPy Array : \n",arr_3D)
shape_of_arr_3D = np.shape(arr_3D)
print("Shape of arr_3D = ", shape_of_arr_3D)
```

#### Python NumPy array shape using shape attribute

```python
np.shape
shape_of_arr_1D = arr_1D.shape
print("Shape of 1D array =", shape_of_arr_1D)
```
