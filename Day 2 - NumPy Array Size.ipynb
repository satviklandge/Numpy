## NumPy Array Size â€“ Np.Size() 
    In NumPy, the np.size() function is used to find the total number of elements in an array.
    It tells you how many values are present in the array, regardless of its shape (1D, 2D, or multi-dimensional).
    
    ðŸ”¹ Key Points
    - np.size() returns total elements
    - Works for all dimensions
    - Same as array.size attribute

```python
import numpy as np
```

```python
array_1 = np.array([[0, 1, 1], [1, 0, 1], [1, 1, 0],[0, 0, 1]])
```

```python
array_1
```

```python
array_1.size
```

```python
arr_axis = np.size(array_1)
```

```python
arr_axis
```

```python
print("NUmber of Rows :",np.size(array_1,axis=0))
print("NUmber of Cols :",np.size(array_1,axis=1))    
```
